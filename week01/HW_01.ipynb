## Question 2: Wide data isn't usually helpful.

```python
df
```

```python
s2 = df.copy()
```

```python
new = pd.melt(s2,id_vars=['country','year'], value_vars=s2.columns[2:58],var_name='diagnosis method', value_name='number of cases')
```

```python
new
```

```python
Data = new.copy()
```

```python
Data
```

```python
Data = Data.replace(regex={r'^new_sp_.([\d]{2}|[\d]{3}|[\d]{4}])$': 'Pulmonary Smear (+)', r'^new_sn_.([\d]{2}|[\d]{3}|[\d]{4}])$': 'Pulmonary Smear (-)',r'^new_ep_.([\d]{2}|[\d]{3}|[\d]{4}])$':'Extrapulmonary',r'^newrel_.([\d]{2}|[\d]{3}|[\d]{4}])$':'Relapse'})
```

```python
Data
```

```python
Data['gender'] = new['diagnosis method']
Data = Data.replace(regex={r'^new_sp_([m])([\d]{2}|[\d]{3}|[\d]{4})$':'male',r'^new_sp_([f])([\d]{2}|[\d]{3}|[\d]{4})$':'female',r'^new_ep_([m])([\d]{2}|[\d]{3}|[\d]{4})$':'male',r'^new_ep_([f])([\d]{2}|[\d]{3}|[\d]{4})$':'female',r'^newrel_[m]([\d]{2}|[\d]{3}|[\d]{4}])$':'male',r'^newrel_[f]([\d]{2}|[\d]{3}|[\d]{4}])$':'female'})
```

```python
Data
```

```python
Data['age'] = new['diagnosis method']
```

```python
Data
```

```python
Data['age'] = Data['age'].str[-4:]
Data['LL'] = Data['age'].str[-4:-2]
Data['UL'] = Data['age'].str[-2:]
Data['LL'] = Data['LL'].replace(['m0','f0'], '0')
Data['LL'] = Data['LL'].replace(['_m','_f'], '65')
```

```python
Data['LL'].unique()
```

```python
Data['UL'].unique()
```

```python
Data["age"] = Data['LL'] +"-"+ Data["UL"]
Data['age'] = Data['age'].replace(['65-65'], '65')
```

```python
Data
```

```python
Data.drop(['LL','UL'], axis=1)
```

```python
Final_Data = Data[['country','year','diagnosis method','gender','age','number of cases']]
```

```python
Final_Data
```

```python
w = Final_Data.copy()
```

```python
w = w[w['number of cases']>=0]
```

```python
w
```

```python
w.groupby('country')['year'].idxmin()
```

```python
Data = w[['country','year','number of cases']].loc[w.groupby('country')['year'].idxmin()]
```

```python
Data
```

# Question 3: Create a graph that shows when countries started to report TB cases.

```python
plt.figure(figsize=(35,15))
plt.xlabel('year')
plt.ylabel("number of cases")
plt.title('Countries started to report TB cases')
plt.plot(Data['country'],Data['year'],'s-g');
plt.tick_params(axis='x', rotation=90)
```
